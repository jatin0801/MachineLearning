```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mlr.csv')
df
```

```python
x= df.iloc[:,:-1].values
y= df.iloc[:,-1].values
```

```python
df.info()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])],remainder='passthrough')
x = ct.fit_transform(x)
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
y_pred= lr.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test , y_pred)
print('mse= ',mse)
print('rmse= ',rmse)
print('mae= ',mae)
print('r2 score= ',r2)
```

**BACKWARD ELIMINATION**

```python
import statsmodels.api as sm
X = x[:,1:]
print(X)
```

```python
constt = np.ones((50,1))
constt
```

```python
X = np.append(arr = constt , values= X, axis = 1)
print(X)
```

```python
x_opt = np.array(X[:,[0,1,2,3,4,5]],dtype=float)
stat = sm.OLS(endog=y,exog=x_opt).fit()
stat.summary()
```

```python
x_opt = np.array(X[:,[0,1,3,4,5]],dtype=float)
stat = sm.OLS(endog=y,exog=x_opt).fit()
stat.summary()
```

```python
x_opt = np.array(X[:,[0,3,4,5]],dtype=float)
stat = sm.OLS(endog=y,exog=x_opt).fit()
stat.summary()
```

```python
x_opt = np.array(X[:,[0,3,5]],dtype=float)
stat = sm.OLS(endog=y,exog=x_opt).fit()
stat.summary()
```

```python
x_opt = np.array(X[:,[0,3]],dtype=float)
stat = sm.OLS(endog=y,exog=x_opt).fit()
stat.summary()
```

```python
print(x_opt)
```

```python
xopt_train,xopt_test,yopt_train,yopt_test = train_test_split(x_opt,y,test_size=0.25,random_state=0)
```

```python
lr_opt = LinearRegression()
lr_opt.fit(xopt_train,yopt_train)
```

```python
yopt_pred = lr_opt.predict(xopt_test)
print(mean_squared_error(yopt_test,yopt_pred))
print(r2_score(yopt_test,yopt_pred))
```

```python
#3d model
from mpl_toolkits import mplot3d
ax = plt.figure().add_subplot(projection='3d')
ax.scatter3D(xopt_test[:,0],xopt_test[:,1],yopt_test,c='red')
ax.plot3D(xopt_test[:,0],xopt_test[:,1],yopt_pred,c='blue')
ax.set_xlabel('constant')
ax.set_ylabel('RnD')
ax.set_zlabel('profit')
plt.show
```
