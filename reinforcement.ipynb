**UPPER CONFIDENCE BOUND**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('reinforcement.csv')
df
```

```python
import math as ma
N = 10000
d = 10
number_of_selection = [0]*10
sum_of_rewards = [0]*d
total_reward = 0
ads_selected = []
for n in range (0,N):
  ad = 0
  max_upper_bound = 0
  for i in range (0,d):
    if number_of_selection[i]>0:
      average_reward = sum_of_rewards[i]/number_of_selection[i]
      delta_i = ma.sqrt(1.5*ma.log(n+1)/number_of_selection[i])
      upper_bound = average_reward + delta_i
    else:
      upper_bound = 1e400
    if upper_bound > max_upper_bound:
      max_upper_bound = upper_bound
      ad = i
  reward = df.values[n,ad]
  ads_selected.append(ad)
  number_of_selection[ad] = number_of_selection[ad]+1
  sum_of_rewards[ad] = sum_of_rewards[ad] + reward
  total_reward = total_reward + reward
```

```python
print(ads_selected)
```

```python
plt.hist(ads_selected)
plt.show()
```
