```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('randomforest.csv')
df
```

```python
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
x
```

```python
y
```

```python
#import and train rf algo
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=9, random_state=0)
rfr.fit(x,y)
```

```python
#creating test np array and predict ans
c= np.array([[4.5],[6.5]])
y_pred = rfr.predict(c)
y_pred
```

```python
x_grid = np.arange(min(x),max(x),0.001)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(x,y,c='red')
plt.plot(x_grid,rfr.predict(x_grid),c='blue')
plt.title("Level vs Salary Graph")
plt.xlabel('Label')
plt.ylabel('Salary')
plt.show()
```
